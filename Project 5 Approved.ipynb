## 3.2 Platform Sales Analysis Over Time

Now let's analyze how sales vary across platforms and years:

````python
# Calculate total sales by platform and year
````

````python
platform_releases = next_gen.groupby(['platform', 'year_of_release']).size().reset_index(name='count')

platform_releases_sorted = platform_releases.sort_values(by=['count', 'year_of_release'], ascending=[False, True])

display(platform_releases_sorted.head(9))
````

````python
# Create a heatmap of platform sales over time
````

````python
platform_sales_pivot = platform_releases_sorted.pivot(index='platform', columns='year_of_release', values='count').fillna(0)

plt.figure(figsize=(16, 10))
sns.heatmap(platform_sales_pivot, cmap='viridis', annot=False, fmt='.0f', linewidths=0.5)

plt.title('Heatmap of Platform Games Released Over Time', fontsize=16)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Platform', fontsize=12)

plt.show()
````

````python
# Identify platforms with declining sales
````

````python
platform_sales = next_gen.groupby(['platform', 'year_of_release'])['total_sales'].sum().reset_index()
platform_totals = platform_sales.groupby('platform')['total_sales'].sum()
valid_platforms = platform_totals[platform_totals >= 70].index
platform_sales = platform_sales[platform_sales['platform'].isin(valid_platforms)]

platform_sales_pivot = platform_sales.pivot(index='year_of_release', columns='platform', values='total_sales').fillna(0)

plt.figure(figsize=(16, 10))
sns.lineplot(data=platform_sales_pivot, dashes=False, palette="tab10")  # Use 'tab10' for better distinction

plt.title('Platform Sales Trends Over Time)', fontsize=16)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Total Sales (in millions)', fontsize=12)

plt.legend(title='Platform', fontsize=10, loc='upper left')
plt.show()
````

# Step 5 : Hypothesis Tests

‚ÄîAverage user ratings of the Xbox One and PC platforms are the same.

‚ÄîAverage user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

‚ÄîHow you formulated the null and alternative hypotheses

<div class="alert alert-success" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">
<h2> Reviewer's comment 2</h2>

    
Good. The null hypothesis always includes the equality sign. This is because the test does not understand exactly how we set the problem: we can say, for instance, let's make sure that they are not equal. Or that they are equal. And regardless of the purpose of our study, we always put equality at null hypothesis. Then the test result is interpreted correctly.

</div>   

````python
from scipy import stats

xbox_ratings = recent_regional_study[recent_regional_study['platform'] == 'XOne']['user_score'].astype(float)
pc_ratings = recent_regional_study[recent_regional_study['platform'] == 'PC']['user_score'].astype(float)

t_stat, p_value = stats.ttest_ind(xbox_ratings, pc_ratings, nan_policy='omit')

if p_value < 0.09:
    print("Reject H‚ÇÄ (Null Hypothesis): The average user ratings for Xbox One and PC platforms are not the same.")
else:
    print("Fail to reject H‚ÇÄ (Null Hypothesis): The average user ratings for Xbox One and PC platforms are the same.")
````

<div class="alert alert-block alert-info">
<b> Explanation! </b> 
    
* This T-Test is used to compare the means of two independent platforms by their ratings. The test is to see if there is a 9% difference between the user score on an Xbox and on PC. The test proved that the ratings are the same with no difference. 
    (I use 9% due to it shows larger differences in the ratings as well as it is my favorite number)
